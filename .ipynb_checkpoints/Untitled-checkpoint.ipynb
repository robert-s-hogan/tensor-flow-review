```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Path to your image
image_path = 'darth-maul-removebg-preview.png'

# Load and decode the image as RGB (3 channels)
image_rgb = tf.io.read_file(image_path)
image_rgb = tf.image.decode_jpeg(image_rgb, channels=3)
image_rgb = tf.image.convert_image_dtype(image_rgb, tf.float32)

# Load and decode the image as RGBA (4 channels)
image_rgba = tf.io.read_file(image_path)
image_rgba = tf.image.decode_jpeg(image_rgba, channels=4)
image_rgba = tf.image.convert_image_dtype(image_rgba, tf.float32)

# Load and decode the image as Grayscale (1 channel)
image_gray = tf.io.read_file(image_path)
image_gray = tf.image.decode_jpeg(image_gray, channels=1)
image_gray = tf.image.convert_image_dtype(image_gray, tf.float32)

# Print shapes
print("RGB Image shape:", image_rgb.shape)
print("RGBA Image shape:", image_rgba.shape)
print("Grayscale Image shape:", image_gray.shape)

# Visualize the images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title('RGB Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image_rgba)
plt.title('RGBA Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(tf.squeeze(image_gray), cmap='gray')  # tf.squeeze removes the single channel dimension
plt.title('Grayscale Image')
plt.axis('off')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the RGB value for black
black_rgb = (0, 0, 0)

# Normalize the RGB value
normalized_black_rgb = [val / 255.0 for val in black_rgb]

# Create an image of 100x100 pixels with the black background
image_black_background = np.ones((100, 100, 3)) * normalized_black_rgb

# Display the image with a black background
plt.imshow(image_black_background)
plt.title('Black Background (RGB: (0, 0, 0))')
plt.axis('off')
plt.show()

# Define the RGB value for white
white_rgb = (255, 255, 255)

# Normalize the RGB value
normalized_white_rgb = [val / 255.0 for val in white_rgb]

# Create an image of 100x100 pixels with the white background
image_white_background = np.ones((100, 100, 3)) * normalized_white_rgb

# Display the image with a white background
plt.imshow(image_white_background)
plt.title('White Background (RGB: (255, 255, 255))')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to display an image with a given RGB background color
def display_background(rgb_value, title):
    normalized_rgb = [val / 255.0 for val in rgb_value]
    image = np.ones((100, 100, 3)) * normalized_rgb
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Common background colors with their RGB values
background_colors = {
    "Black": (0, 0, 0),
    "White": (255, 255, 255),
    "Gray": (128, 128, 128),
    "Red": (255, 0, 0),
    "Green": (0, 255, 0),
    "Blue": (0, 0, 255)
}

# Display each background color
for color_name, rgb_value in background_colors.items():
    display_background(rgb_value, f"{color_name} Background (RGB: {rgb_value})")
```
